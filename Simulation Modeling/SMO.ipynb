```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('ggplot')
```

```python
k = 10 #number of requirements

def queueingSystem(k):
    total_WT = 0   #total waiting time
    total_DT = 0  #total downtime

    dt = 0 
    wt = 0
    
    i = 0
    
    while i <= k:
        t = np.random.uniform(1, 15)
        t = t - wt
        s = np.random.uniform(5, 7)
        
        if s > t:
            dt = 0
            wt = s - t
            
            total_WT += wt
            
        elif s < t:
            wt = 0
            dt = t - s
            
            total_DT += dt
            
        else:
            dt = 0
            wt = 0
        
        i += 1
        
    return total_WT/k, total_DT/k    
```

```python
array = []
for i in range(100):
    array.append(queueingSystem(20))
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

ax[0].scatter(np.arange(len(array)), np.array(array).T[0], s=15)
ax[1].scatter(np.arange(len(array)), np.array(array).T[1], s=15)

ax[0].set_title('Mean Total Waiting Time')
ax[1].set_title('Mean Total DownTime')

plt.subplots_adjust(hspace=0.2)
plt.show()
```
