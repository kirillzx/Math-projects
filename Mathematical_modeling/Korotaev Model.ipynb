```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
%matplotlib inline

plt.style.use('ggplot')
```

```python
from scipy.integrate import odeint, RK45, solve_ivp
```

```python
def korotaev_system(values, t, a, b, c):
    N, T, E = values
    
    return [
        a * T * N * (1-E),
        b * T * E,
        c * T * E * (1-E)
    ]
```

```python
a = 1.2
b = 0.2
c = 1.01

N0 = 2
T0 = 1
E0 = 1

t = np.linspace(0, 10, 100)

solution = odeint(korotaev_system, [N0, T0, E0], t, args=(a, b, c))
```

```python
N, T, E = solution[:, 0], solution[:, 1], solution[:, 2]
```

```python
Y = 0.8*N*T
```

```python
plt.subplots(figsize=(10, 5), dpi=100)

plt.plot(t, N, label='N')
plt.plot(t, T, label='T')
plt.plot(t, E, label='E')
plt.plot(t, Y, label='Y')

plt.legend()
plt.title('Korotaev Model', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.show()
```
