```python
import numpy as np
import matplotlib.pyplot as plt
import math
plt.style.use("ggplot")
```

```python
#Runge-Kutt method
def runge_Kutt(f, g, t0, x0, y0, h, b):
    t = t0
    x = x0
    y = y0
    arrayX = np.array([])
    arrayY = np.array([])
    arrayT = np.array([])
    while t<b:
        k1 = g(t, x, y)
        q1 = f(t, x, y)
        
        k2 = g(t + h/2, x + (h*q1)/2, y + (h*k1)/2)
        q2 = f(t + h/2, x + (h*q1)/2, y + (h*k1)/2)
        
        k3 = g(t + h/2, x + (h*q2)/2, y + (h*k2)/2)
        q3 = f(t + h/2, x + (h*q2)/2, y + (h*k2)/2)
        
        k4 = g(t + h, x + h*q3, y + h*k3)
        q4 = f(t + h, x + h*q3, y + h*k3)
        
        y = y + h*(k1 + 2*k2 + 2*k3 + k4)/6
        x = x + h*(q1 + 2*q2 + 2*q3 + q4)/6
        
        arrayX = np.append(arrayX, x)
        arrayT = np.append(arrayT, t)
        arrayY = np.append(arrayY, y)
        
        t += h
    return arrayT, arrayX, arrayY
```

```python
def fx(t, x, y):
    return x - y - x*(x**2+y**2)

def fy(t, x, y):
    return x + y - y*(x**2+y**2)

def fxl(t, x, y):
    return a1*math.log(x) - b11*math.log(x)**2 - b12*math.log(x)*math.log(y)

def fyl(t, x, y):
    return a2*math.log(y) - b21*math.log(x)*math.log(y) - b22*math.log(y)**2

def fxp(t, r, phi):
    return r*(1 - r**2)

def fyp(t, r, phi):
    return 1

p1, p2, p3 = runge_Kutt(fxp, fyp, 0, 0.2, 0.4, 0.01, 100)
print(f"Solution: x = {p2[-1]}, y = {p3[-1]}")
```

```python
fig, ax = plt.subplots(figsize=(11,8))
n = len(p1)
t = np.linspace(0, 100, n)
t1 = np.linspace(0, 14, 20)
plt.plot(t, p2, label="X")
plt.plot(t, p3, label="Y")

plt.legend(prop={'size': 12})
plt.show()
```

## r = r(1-r**2)

```python
fig, ax = plt.subplots(figsize=(12,9))
for i in range(5, 10):
    p1, p2, p3 = runge_Kutt(fxp, fyp, 0, math.cos(i*math.pi/20), 1, 0.01, 100)
    plt.polar(p3, p2, c='purple')
    
for i in range(1, 6):
    p1, p2, p3 = runge_Kutt(fxp, fyp, 0, i/2, 1, 0.01, 100)
    plt.polar(p3, p2, c='purple')    

# plt.legend(prop={'size': 12})
plt.show()
```
